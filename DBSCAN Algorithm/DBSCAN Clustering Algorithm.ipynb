# DBSCAN Algorithm- ML Model Training

```python
from sklearn.cluster import DBSCAN
```

```python
dbn = DBSCAN(eps=3,min_samples=4)
```

```python
X = df.iloc[:,[3,4]].values
dbn.fit_predict(X)
labels = dbn.labels_
labels
```

```python
# Number of clusters in labels, ignoring noise if present. 
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0) 
```

```python
df['cluster'] = y_pred
df.head()
```

```python
# Now we will plot graph for all 9 clusters
plt.figure(figsize=(16,12))
import numpy as np
i=0
df_new = []
c_rng = np.arange(0,no_clusters)
colours = ['b','g','r','c','m','y','k','orange','purple']
for i in c_rng :
    df_new.append( df[df['cluster']==i] )
    #     print(df_new[i].head())
    plt.scatter(df_new[i]['Annual Income (k$)'],df_new[i]['Spending Score (1-100)'],color = colours[i])
    
```
