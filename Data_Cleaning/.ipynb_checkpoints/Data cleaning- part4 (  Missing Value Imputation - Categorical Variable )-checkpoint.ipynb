```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ignore warnings
import warnings
warnings.filterwarnings("ignore")

# show all rows and all columns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df = pd.read_csv(r"C:\Users\user\Pictures\PYTHON_PANDAS\archive\housetrain.csv")
df
```

```python
cat_vars = df.select_dtypes(include = 'object')
cat_vars.head()
```

```python
miss_val_per = cat_vars.isnull().mean()*100
miss_val_per
```

```python
miss_val_per[miss_val_per>20].keys()
```

```python
drop_vars = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
cat_vars.drop(columns = drop_vars,axis=1,inplace=True)
```

```python
cat_vars.shape
```

```python
isnull_per = cat_vars.isnull().mean()*100
miss_var = isnull_per[isnull_per>0].keys()
miss_var
```

```python
cat_vars['MasVnrType'].mode()
```

```python
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0])
```

```python
for var in miss_var :
    cat_vars[var].fillna(cat_vars[var].mode()[0], inplace=True)
    print(var,"=",cat_vars[var].mode()[0])
    
```

```python
cat_vars.isnull().sum().sum()
```

```python
#data distribution
plt.figure(figsize=(16,9))
for i,var in enumerate(miss_var) :
    plt.subplot(4,3,i+1)
    plt.hist(cat_vars[var], label="Imput")
    plt.hist(df[var].dropna(), label="Original")
    plt.legend()
```

```python
# update dataframe
df.update(cat_vars)
df.drop(columns=drop_vars,inplace=True)
```
