# --------(2) using Scikit-Learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse=False)
```

```python
banglore_df['size'].fillna(banglore_df['size'].mode()[0], inplace=True)
```

```python
oh_enc_arr = oh_enc.fit_transform(banglore_df[['area_type']])
oh_enc_arr      # numpy array
```

```python
#convert into dataframe
oh_enc_df = pd.DataFrame(oh_enc_arr,columns=['area_type_Built-up Area','area_type_Carpet Area','area_type_Plot Area','area_type_Super built-up Area'])
```
