```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = pd.read_csv(r"C:\Users\user\Pictures\PYTHON_PANDAS\tips.csv")
tips_df
```

```python
# show all rows and all columns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
null_var = tips_df.isnull().sum() / tips_df.shape[0]*100
null_var
```

```python
tips_df.isnull().sum().sum()
```

```python
df = pd.get_dummies(tips_df)
df
```

```python
X = df.drop('total_bill',axis=1)
y = df['total_bill']   # price is a dependent variable

print("Shape of X = ", X.shape)
print("Shape of y = ", y.shape)
```

# Split Data

```python
# now we will split data into train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=51)
print("Shape of X_train = ", X_train.shape)
print("Shape of y_train = ", y_train.shape)
print("Shape of X_test = ", X_test.shape)
print("Shape of y_test = ", y_test.shape)
```

# Feature Scaling

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

# Linear Regression - ML Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train,y_train)
```

```python
#till now our model has learnt the coffiecients

lr.coef_                            # all slopes (B1,B2,......,beta n)
```

```python
lr.intercept_             # intercept value
```

```python
lr.predict(X_test)          # prediction of all houses that are in X_test data
```

```python
y_test
```

```python
lr.score(X_test, y_test)
```

# Implementing Ridge and Lasso Regression

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
rd = Ridge()

rd.fit(X_train,y_train)

rd.score(X_test,y_test)
```

```python
ls = Lasso()

ls.fit(X_train,y_train)

ls.score(X_test,y_test)
```

```python
rd2 = Ridge(alpha=0.5)

rd2.fit(X_train,y_train)

rd2.score(X_test,y_test)
```

```python
ls2 = Lasso(alpha=0.5)

ls2.fit(X_train,y_train)

ls2.score(X_test,y_test)
```
