```python
import pandas as pd
```

```python
#ignore warnings
import warnings
warnings.filterwarnings("ignore")

# show all rows and all columns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

# data preprocessing
cleaning the data

```python
banglore_df = pd.read_csv(r"C:\Users\user\Pictures\PYTHON_PANDAS\banglore.csv")
banglore_df.shape
```

```python
null_var = banglore_df.isnull().sum()/banglore_df.shape[0] *100
null_var
```

```python
drop_columns = null_var[null_var>10].keys()    # more then 10% null columns will be dropped
drop_columns
```

```python
df2_drop_clm = banglore_df.drop(columns = drop_columns)
df2_drop_clm.shape
```

```python
df3_drop_rows = df2_drop_clm.dropna()
```

```python
df3_drop_rows.isnull().sum()       # data is cleaned
```

```python
df3_drop_rows.head()
```

# OneHotEncoding
Now change categorical data to numerical data


```python
df = pd.get_dummies(df3_drop_rows)
df.head()
```

```python
df.shape
```

# Split Data

```python
X = df.drop('price',axis=1)
y = df['price']   # price is a dependent variable

print("Shape of X = ", X.shape)
print("Shape of y = ", y.shape)
```

```python
# now we will split data into train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=51)
print("Shape of X_train = ", X_train.shape)
print("Shape of y_train = ", y_train.shape)
print("Shape of X_test = ", X_test.shape)
print("Shape of y_test = ", y_test.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

# Linear Regression - ML Model Training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train,y_train)
```

```python
#till now our model has learnt the coffiecients

lr.coef_                            # all slopes (B1,B2,......,beta n)
```

```python
lr.intercept_             # intercept value
```

# Predict the Value of Home and Test

```python
X_test[0,:]            # 0th house and all columns
```

```python
lr.predict([X_test[1,:]])     # prediction of 1th house
```

```python
lr.predict(X_test)          # prediction of all houses that are in X_test data
```

```python
y_test
```

```python
lr.score(X_test, y_test)
```

# Implement Ridge and Lasso Rigression

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
rd = Ridge()

rd.fit(X_train,y_train)

rd.score(X_test,y_test)
```

```python
ls = Lasso()

ls.fit(X_train,y_train)

ls.score(X_test,y_test)
```

```python
ls2 = Lasso(alpha=0.5)

ls2.fit(X_train,y_train)

ls2.score(X_test,y_test)
```

```python
rd2 = Ridge(alpha=2)

rd2.fit(X_train,y_train)

rd2.score(X_test,y_test)
```
